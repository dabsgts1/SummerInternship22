```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.tree import DecisionTreeRegressor
```

```python
# save filepath to variable for easier access
home_data_file_path = 'train.csv'

# read the data and store data in DataFrame titled home_data
home_data = pd.read_csv(home_data_file_path) 

# print a summary of the data in Melbourne data
home_data.describe()
```

```python
# printing all columns
home_data.columns
```

```python
# ask for features the user wants considered
features = input('Enter features to be considered separated by a comma: ')
ft_list = []
features = features.replace(' ','')
features = features.replace("'","")

# converting input to a list and confirming that each feature is valid
for feature in features.split(','):
    if feature in home_data.columns:
        ft_list.append(feature)
        
# 'Fence', 'MiscFeature', 'MiscVal', 'MoSold', 'YrSold', 'SaleType',        'SaleCondition'
print(ft_list)

train_X = home_data[ft_list] # list of features from our training set train.csv
train_X = pd.get_dummies(train_X)
train_y = home_data.SalePrice # corresponding Sale Price

test_path = 'test.csv'
test_data = pd.read_csv(test_path) 
test_X = test_data[ft_list] # list of features from our testing set test.csv
test_X = pd.get_dummies(test_X)
```

```python
pos = 0
for col in train_X.columns:
    if not col in test_X.columns:
        test_X.insert(pos, col, 0)
    pos = pos + 1
```

```python
train_X.columns
```

```python
test_X.columns
```

```python
home_model = DecisionTreeRegressor(random_state=1)
home_model.fit(train_X, train_y)


print("Making predictions for the following houses:")
test_X
```

```python
print("The predictions are")
home_predictions = home_model.predict(test_X)
home_predictions
```

```python
ind = 1461
df = pd.DataFrame(columns=['Id','SalePrice'])
for price in home_predictions:
    newrow = pd.DataFrame({'Id': [ind], 'SalePrice':[price]})
    df = pd.concat([df, newrow], ignore_index = True)
    ind = ind + 1
    
df
```

```python
df.to_csv (r'house_prices.csv', index=False, header=True)
```
